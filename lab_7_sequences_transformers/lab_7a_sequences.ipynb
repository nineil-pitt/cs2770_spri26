## Promt engineering - Instruction Tuning
FLAN-T5 Example. It was adapted from [here](https://huggingface.co/docs/transformers/v4.26.1/en/model_doc/flan-t5).

```python
# Model Loading
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
```

```python
# Prediction
inputs = tokenizer("A step by step recipe to make bolognese pasta:", return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))
```

### Helper Functions

```python
from IPython.display import HTML
from base64 import b64encode
import os

def convert_avi_to_mp4(avi_file_path, output_name):
  os.popen("ffmpeg -i '{input}' -ac 2 -b:v 2000k -c:a aac -c:v libx264 -b:a 160k -vprofile high -bf 0 -strict experimental -f mp4 '{output}.mp4'".format(input = avi_file_path, output = output_name))
  return True

def play(filename):
  html = ''
  video = open(filename,'rb').read()
  src = 'data:video/mp4;base64,' + b64encode(video).decode()
  html += '<video width=1000 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src
  return HTML(html)
```

```python
from torchvision.io.video import read_video
from torchvision.models.video import r3d_18, R3D_18_Weights

VIDEO_CLASSIFICATION = "./test/v_SoccerJuggling_g23_c01.avi"
vid, _, _ = read_video(VIDEO_CLASSIFICATION, output_format="TCHW") # If error, see comment from "pip install av" command
# vid = vid[:32]  # optionally shorten duration

# Step 1: Initialize model with the best available weights
weights = R3D_18_Weights.DEFAULT
model = r3d_18(weights=weights)
model.eval()

# Step 2: Initialize the inference transforms
preprocess = weights.transforms()

# Step 3: Apply inference preprocessing transforms
batch = preprocess(vid).unsqueeze(0)

# Step 4: Use the model and print the predicted category
prediction = model(batch).squeeze(0).softmax(0)
label = prediction.argmax().item()
score = prediction[label].item()
category_name = weights.meta["categories"][label]
print(f"{category_name}: {100 * score}%")
```
